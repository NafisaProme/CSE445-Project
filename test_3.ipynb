```python
from sklearn.ensemble import GradientBoostingRegressor
import random

# Assume X and y are your feature matrix and target variable respectively
# Update X and y with the relevant data from your dataset

# Create a list to store the individual models
models = []

# Train ten Random Forest models and add them to the list
for _ in range(30):
    # Create a Random Forest model
    r_state = random.randint(0, 45)
    rf = GradientBoostingRegressor(random_state=r_state)

    # Train the model on the data
    rf.fit(X, y)

    # Add the trained model to the list
    models.append(rf)

# Make predictions using each individual model
predictions = []
for model in models:
    pred = model.predict(X)
    predictions.append(pred)

# Combine the predictions using the average
ensemble_pred = np.mean(predictions, axis=0)

# Evaluate the ensemble model
mse = mean_squared_error(y, ensemble_pred)
mae = mean_absolute_error(y, ensemble_pred)
r2 = r2_score(y, ensemble_pred)

# Print the evaluation metrics
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)
```

```python
for i, (prediction, actual) in enumerate(zip(ensemble_pred,y), start=1):
    print(f"Data Point {i}: Predicted Price: {prediction:.2f}, Actual Price: {actual:.2f}")
```

```python
# Using emsemble classifier 
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import VotingRegressor

reg1 = GradientBoostingRegressor(random_state=41)
reg2 = RandomForestRegressor(random_state=40)
reg3 = LinearRegression()
reg4 = GradientBoostingRegressor(random_state=35)

voting_regressor = VotingRegressor(estimators=[('gb', reg1), ('rf', reg2), ('lr', reg3), ('gb2', reg4)])
voting_regressor.fit(X_train, y_train)
y_pred = voting_regressor.predict(X_test)

# Evaluate the ensemble model
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)
```

```python
import xgboost as xgb
import random

b = []
lim1 = 10
lim2 = 20
lim = 2 * lim1
for i in range(0, lim):
    b.append(str(i))
    
r_state = random.randint(0, 45)
# booster = GradientBoostingRegressor(random_state=r_state)
    
b_m = dict(zip(b, ["NULL"] * len(b)))

for i in range(0, lim1):
    r_state = random.randint(0, 45)
    booster = GradientBoostingRegressor(random_state=r_state)
    b_m[b[i]] = booster
    
for i in range(lim1, lim2):
    r_state = random.randint(0, 45)
    booster = RandomForestRegressor(random_state=r_state)
    b_m[b[i]] = booster

reg = {}
reg[0] = xgb.XGBRegressor(random_state=27)
reg[1] = xgb.XGBRegressor(random_state=32)
reg[2] = xgb.XGBRegressor(random_state=35)
reg[3] = xgb.XGBRegressor(random_state=41)
reg[4] = GradientBoostingRegressor(random_state=42)

voting_regressor = VotingRegressor(estimators=[
    (b[0], reg[0]),
    (b[1], reg[1]),
    (b[2], reg[2]),
    (b[3], reg[3]),
    (b[4], reg[4]),
])
voting_regressor.fit(X_train, y_train)
ensemble_pred = voting_regressor.predict(X_test)

# Evaluate the ensemble model
mse = mean_squared_error(y_test, ensemble_pred)
mae = mean_absolute_error(y_test, ensemble_pred)
r2 = r2_score(y_test, ensemble_pred)

# Print the evaluation metrics
print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R-squared:", r2)

for i, (prediction, actual) in enumerate(zip(ensemble_pred,y_test), start=1):
    print(f"Data Point {i}: Predicted Price: {prediction:.2f}, Actual Price: {actual:.2f}")
```
